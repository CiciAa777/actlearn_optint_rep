```python
from argparse import Namespace
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

from optint.data import synthetic_instance
from optint.run import run
```

```python
import networkx as nx


d = nx.DiGraph()
d.add_nodes_from(list(range(5)))
for (i, j) in [(0,1), (0,2), (2,3), (3,4), (1,2), (1,4)]:
	d.add_edge(i, j)
    
color_d = []
for i,j in d.edges:
	color_d.append('cornflowerblue')
    
dict_l = {node: r'$x_{}$'.format(str(node+1)) for node in range(5)}
nw_ax = plt.subplot2grid((4, 4), (0, 0), colspan=12, rowspan=12)
plt.gcf().set_size_inches(4, 4)
pos = nx.circular_layout(d)
nx.draw(d, pos=pos, node_color='w', ax=nw_ax, edge_cmap=plt.cm.Blues, edge_vmin=0, edge_vmax=1, edge_color=color_d, edgecolors='dimgrey', linewidths=3, width=3, node_size=1200, font_size=15)
nx.draw_networkx_labels(d, pos, labels=dict_l, ax=nw_ax, font_size=20)
plt.tight_layout()
plt.savefig('figures/dag1.pdf')
```
